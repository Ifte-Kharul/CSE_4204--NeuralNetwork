```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
X = df.drop(['Outcome'],axis=1)
X.head()
```

```python
#correlation matrix between features
correlation_matrix = df.corr()
```

```python
#Creating a Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#histogram of each features
features_column = X.columns

fig, axes = plt.subplots(nrows=2,ncols=4, figsize= (16,8))
axes = axes.flatten()
for i, feature in enumerate(features_column):
    ax = axes[i]
    ax.hist(df[feature], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(feature)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')


plt.tight_layout()
plt.show()
```

```python
%matplotlib widget
plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
# fg = ax.scatter3D(X['Glucose'],X['BMI'],X['Pregnancies'])
fg = ax.scatter3D(X['Glucose'],X['BMI'],df['Age'],c=df['Outcome'])
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
plt.show()
```
